I want to look at the performance of sellers in their first month and see what type of sellers are successful and what things they have done to get there.

The difficulty here is finding stats that we can limit by time. We will want to only look at stats between Booth.created_at + 30 days: https://www.bonanza.com/admin/bonanza_models/booth
