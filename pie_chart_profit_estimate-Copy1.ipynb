```python
sum_values = sum([category.profit_estimate[0] for category in categories])
```

```python


values = [(category.profit_estimate[0]/sum_values) *100 for category in categories]
print(values)

labels = top_categories
```

```python
fig = {
  "data": [
    {
      "values": values,
      "labels": labels,
      "marker": {'colors': ['rgb(33, 75, 99)',
                                  'rgb(79, 129, 102)',
                                  'rgb(151, 179, 100)',
                                  'rgb(175, 49, 35)',
                                  'rgb(36, 73, 147)']},
      "name": "values",
      "hoverinfo":"label+percent+name",
      "hole": .4,
      "type": "pie"
    }],
  "layout": {
        "title":"Mean Profit",
        "annotations": [
            {
                "font": {
                    "size": 20
                },
                "showarrow": False,
                "text": round(sum_values/6),
                
            }
        ]
    }
}
po.plot(fig, filename='donut.html')
```
